```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


current_dir = '/home/yshwang/algorithm/image_processing'
img = cv2.imread(current_dir + '/image/milky-way-g347b5618f_1280.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(img)

img.shape
```

```python
img
```

```python
img_ = cv2.rectangle(img.copy(), (1024, 768), (1280, 1024), color = (0, 255, 0), thickness = 3)
```

```python
plt.imshow(img_)
```

```python
height, width = img.shape[0], img.shape[1]

def GCD(x, y):
    while(y):
        x, y = y, x%y
    return x

size = GCD(height, width)
size
```

```python
points = []

for i in range(width//size):
    for j in range(height//size):
        points.append((i*size, j*size))
```

```python
len(points), points
```

```python
img_ = img.copy()

for x, y in points:
    img_ = cv2.rectangle(img_, (x, y), (x+size, y+size), color = (0, 255, 0), thickness = 2)
```

```python

plt.figure(figsize=(15, 15))
plt.imshow(img_)
```
