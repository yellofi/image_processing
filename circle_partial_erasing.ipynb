```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.ones((256, 256, 3))

center_coordinate = (256//2 -1, 256//2 - 1)
radius = 100
color = (255, 255, 0)

thickness = -1

circle_on_img = cv2.circle(image.astype('uint8')*255, center_coordinate, radius, color, thickness)
label = cv2.cvtColor(circle_on_img.astype('uint8'), cv2.COLOR_BGR2GRAY)
label = np.invert(label)

plt.figure(figsize = (15, 10))
# plt.subplot(1, 2, 1)
# plt.imshow(image)
plt.subplot(1, 2, 1)
plt.title('a circle on white background')
plt.imshow(circle_on_img)
plt.subplot(1, 2, 2)
plt.title('annoataion on a circle')
plt.imshow(label, cmap = 'gray')
plt.show()
```

```python
injection_on_img = np.ones((128, 128, 3))
injection_on_label = np.zeros((128, 128))

circle_on_img_, label_ = circle_on_img.copy(), label.copy()
circle_on_img_[:128, :128, :] = injection_on_img.astype('uint8')*255
label_[:128, :128] = injection_on_label

plt.figure(figsize = (15, 10))
# plt.subplot(1, 2, 1)
# plt.imshow(image)
plt.subplot(1, 2, 1)
plt.title('the circle corrupted due to injection of white square')
plt.imshow(circle_on_img_)
plt.subplot(1, 2, 2)
plt.title('the annotation for the image corrupted as well')
plt.imshow(label_, cmap = 'gray')
plt.show()
```
